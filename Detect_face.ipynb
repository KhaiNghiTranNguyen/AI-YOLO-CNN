```python
import os
import cv2
import numpy as np
from mtcnn.mtcnn import MTCNN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import files
import kagglehub

# Tải dataset Avengers Face Recognition
dataset_path = kagglehub.dataset_download("rawatjitesh/avengers-face-recognition")
print("Path to dataset files:", dataset_path)

# Kiểm tra cấu trúc thư mục chi tiết
print("\nCấu trúc thư mục dataset:")
for root, dirs, files in os.walk(dataset_path):
    print(f"Thư mục: {root}")
    print(f"Các thư mục con: {dirs}")
    print(f"Số file: {len(files)}")
    if files:
        print(f"Ví dụ file: {files[:2]}")
    print("-" * 50)

# Đường dẫn tới thư mục chứa ảnh
data_dir = os.path.join(dataset_path, 'cropped_images')  # Cập nhật thành 'cropped_images'
img_size = (160, 160)  # Kích thước ảnh đầu vào cho CNN

# Kiểm tra một ảnh mẫu để xác minh MTCNN
sample_img_path = os.path.join(data_dir, 'chris_evans', 'chris_evans33.png')
print(f"Kiểm tra ảnh mẫu: {sample_img_path}")
img = cv2.imread(sample_img_path)
if img is not None:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.title("Ảnh mẫu")
    plt.show()
    detector = MTCNN()
    faces = detector.detect_faces(img_rgb)
    print(f"Số khuôn mặt phát hiện trong ảnh mẫu: {len(faces)}")
    if len(faces) > 0:
        print("Tọa độ khuôn mặt:", faces[0]['box'])
else:
    print("Không đọc được ảnh mẫu.")

# Hàm tải và xử lý dữ liệu ảnh
def load_data():
    images = []
    labels = []
    detector = MTCNN()

    for person_name in os.listdir(data_dir):
        person_path = os.path.join(data_dir, person_name)
        if os.path.isdir(person_path):
            for img_name in os.listdir(person_path):
                img_path = os.path.join(person_path, img_name)
                img = cv2.imread(img_path)
                if img is None:
                    print(f"Không đọc được ảnh: {img_path}")
                    continue
                # Tiền xử lý: Phóng to và tăng độ sáng
                img = cv2.resize(img, (0, 0), fx=2, fy=2)  # Phóng to gấp đôi
                img = cv2.convertScaleAbs(img, alpha=1.2, beta=30)  # Tăng độ sáng
                img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # Phát hiện khuôn mặt bằng MTCNN
                faces = detector.detect_faces(img_rgb)
                if len(faces) > 0:
                    x, y, w, h = faces[0]['box']
                    # Đảm bảo tọa độ hợp lệ
                    x, y = max(0, x), max(0, y)
                    face = img_rgb[y:y+h, x:x+w]
                    if face.size == 0:
                        print(f"Vùng khuôn mặt rỗng: {img_path}")
                        continue
                    face = cv2.resize(face, img_size)
                    images.append(face)
                    labels.append(person_name)
                else:
                    print(f"Không phát hiện khuôn mặt: {img_path}")

    return np.array(images), np.array(labels)

# Tải dữ liệu
try:
    images, labels = load_data()
    if len(images) == 0:
        raise ValueError("Không tìm thấy khuôn mặt nào trong dataset. Vui lòng kiểm tra ảnh.")
except FileNotFoundError:
    print(f"Thư mục {data_dir} không tồn tại. Vui lòng kiểm tra lại đường dẫn hoặc cấu trúc dataset.")
    raise

# Mã hóa nhãn (chuyển tên thành số)
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(images, encoded_labels, test_size=0.2, random_state=42)

# Chuẩn hóa dữ liệu
X_train = X_train / 255.0
X_test = X_test / 255.0

# Xây dựng mô hình CNN
def build_cnn_model(num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(160, 160, 3)),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Tạo và huấn luyện mô hình
num_classes = len(label_encoder.classes_)
model = build_cnn_model(num_classes)

# Tăng cường dữ liệu
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

# Huấn luyện mô hình
history = model.fit(
    datagen.flow(X_train, y_train, batch_size=32),
    epochs=20,
    validation_data=(X_test, y_test)
)

# Lưu mô hình trong Colab
model.save('/content/face_recognition_model.h5')

# Đánh giá mô hình
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Độ chính xác trên tập kiểm tra: {test_acc:.2f}")
```

```python

from google.colab import files

# Hàm nhận diện khuôn mặt
def recognize_face(image_path, model, label_encoder):
    detector = MTCNN()
    img = cv2.imread(image_path)
    if img is None:
        return "Không đọc được ảnh"
    img = cv2.resize(img, (0, 0), fx=2, fy=2)
    img = cv2.convertScaleAbs(img, alpha=1.2, beta=30)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    faces = detector.detect_faces(img_rgb)
    if len(faces) == 0:
        return "Không tìm thấy khuôn mặt"

    x, y, w, h = faces[0]['box']
    x, y = max(0, x), max(0, y)
    face = img_rgb[y:y+h, x:x+w]
    if face.size == 0:
        return "Vùng khuôn mặt không hợp lệ"
    face = cv2.resize(face, img_size)
    face = face / 255.0
    face = np.expand_dims(face, axis=0)

    prediction = model.predict(face)
    predicted_class = np.argmax(prediction, axis=1)
    predicted_name = label_encoder.inverse_transform(predicted_class)[0]

    return predicted_name

# Vẽ biểu đồ độ chính xác và mất mát
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Độ chính xác huấn luyện')
plt.plot(history.history['val_accuracy'], label='Độ chính xác kiểm tra')
plt.title('Độ chính xác qua các epoch')
plt.xlabel('Epoch')
plt.ylabel('Độ chính xác')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Mất mát huấn luyện')
plt.plot(history.history['val_loss'], label='Mất mát kiểm tra')
plt.title('Mất mát qua các epoch')
plt.xlabel('Epoch')
plt.ylabel('Mất mát')
plt.legend()

plt.savefig('/content/training_history.png')
plt.show()

# Upload ảnh kiểm tra
print("Vui lòng upload ảnh kiểm tra (jpg hoặc png):")
uploaded = colab_files.upload()

# Hiển thị và nhận diện trên ảnh vừa upload
for filename in uploaded.keys():
    test_image_path = filename
    # Hiển thị ảnh
    img = cv2.imread(test_image_path)
    if img is not None:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.title(f"Ảnh upload: {filename}")
        plt.show()
    else:
        print(f"Không đọc được ảnh: {test_image_path}")
        continue

    # Nhận diện
    predicted_name = recognize_face(test_image_path, model, label_encoder)
    print(f"Khuôn mặt được nhận diện trong {filename}: {predicted_name}")

# Tùy chọn: Tải mô hình và biểu đồ về máy
print("Tải mô hình và biểu đồ về máy:")
colab_files.download('/content/face_recognition_model.h5')
colab_files.download('/content/training_history.png')
```
